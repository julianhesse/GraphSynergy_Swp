# Results from removing outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("IQR.csv")
df = df.sort_values(by='IQR', ascending=True).reset_index(drop=True)

# Define metrics and IQR values
metrics = ['accuracy', 'precision', 'recall', 'roc_auc', 'pr_auc', 'f1_score', 'mcc']
iqr_values = df['IQR'].astype(str)
num_metrics = len(metrics)
num_iqr = len(iqr_values)

# Set the width of bars and x-axis positions
bar_width = 0.1  # Adjust bar width for spacing
x = np.arange(num_metrics)  # X-axis positions for metric groups

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars for each IQR value
for i, iqr in enumerate(iqr_values):
    ax.bar(x + i * bar_width, df.loc[i, metrics], width=bar_width, label=f'{iqr}')

# Format the chart
ax.set_xticks(x + (num_iqr / 2) * bar_width)  # Center X-axis labels
ax.set_xticklabels(metrics, rotation=45, ha='right')  # Metric names as labels
ax.set_ylabel('Metric Value')
ax.set_title('Comparison of Metrics Across Different IQR Multipliers')
ax.legend(title='IQR Multiplier')

# Show the plot
plt.tight_layout()
plt.show()
```
