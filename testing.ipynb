### Create dataframe with indices and count

```python
grouped_df = drug_combination_df.groupby(['Drug1', 'Drug2'])
result_data = []
for (col1, col2), group in grouped_df:
    indices = group.index.tolist()
    count = len(indices)
    result_data.append([col1, col2, indices, count])

# Create a new DataFrame from the collected data
result_df = pd.DataFrame(result_data, columns=['Drug1', 'Drug2', 'indices', 'count'])

result_df = result_df.sort_values(by='count', ascending=False).reset_index(drop=True)
```
