```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_path = 'saved/log/DrugCombDB/1206_182941/'
index = 0
file = f'metrics_fold_{index}.csv'

df = pd.read_csv(data_path + file)

df
```

```python
for metric in ['loss','roc_auc', 'pr_auc', 'f1_score', 'mcc']:
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['epoch'], y=df[f'val_{metric}'], mode='lines', name='val_loss'))
    fig.add_trace(go.Scatter(x=df['epoch'], y=df[metric], mode='lines', name='train_loss'))
    fig.update_layout(title=metric)
    fig.show()
```
